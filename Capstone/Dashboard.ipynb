```python
import PySimpleGUI as psg
import pandas as pd
import numpy as np
import warnings
from itertools import chain
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import pyperclip



# settings
warnings.filterwarnings("ignore")


###############################################################################################################
#                                     Global settings / setup                                                 #
###############################################################################################################

df = pd.read_csv('ingredientDB.csv')
name = df['name']

psg.set_options(font=("Arial",12))

allergicFoods = []

ingredientUmbrella = df = pd.read_csv('IngredientDB.csv')

ingredientUmbrella = ['chicken',
                     'salmon',
                     'turkey',
                     'peas',
                     'sweet potatoes',
                     'beet',
                     'flaxseed',
                     'fish meal',
                     'blueberries',
                     'carrot',
                     'kelp',
                     'rosemary',
                     'corn',
                     'rice',
                     'oat',
                     'barley',
                     'tomato',
                     'alfalfa',
                     'yeast',
                     'chicory',
                     'egg',
                     'bison',
                     'venison',
                     'beef',
                     'rasperries',
                     'sorghum',
                     'citric acid',
                     'pork',
                     'salt',
                     'soy',
                     'lamb',
                      'tapioca',
                      'canola',
                      'sunflower',
                      'garbanzo',
                      'pumpkin',
                      'blue',
                      'yellow',
                      'red',
                      'blood meal',
                      'seaweed',
                      'yeast',
                      'green tea',
                      'spearmint',
                      'sugar',
                      'apple',
                      'broccli',
                      'potato',
                      'mussel',
                      'garlic',
                      'millet',
                      'quinoa',
                      'chia',
                      'cheese',
                      'cranberries',
                      'cocounut',
                      'grass',
                      'ginger',
                      'chamomille',
                      'cinnamon',
                      'kale',
                      'oranges',
                      'algae',
                      'bone',
                      'coconut',
                      'krill',
                      'psyllium',
                      'marigold',
                      'paprika',
                      'lentils',
                      'herring',
                      'kelp',
                      'mixed tocopherols',
                      'bacon',
                      'celery',
                      'vinegar',
                      'flavor',
                      'pinto',
                      'gelatin',
                      'collard greens',
                      'pears',
                      'sarsaparilla',
                      'althea',
                      'rose hip',
                      'pollock',
                      'juniper berries',
                      'cellulose',
                      'spelt',
                      'inositol',
                      'banana',
                      'kiwi',
                      'mango',
                      'papaya',
                      'pineapple',
                      'aloe vera',
                      'pecan',
                      'citrus',
                      'rye',
                      'whitefish',
                      'milo',
                      'whiting',
                      'catfish',
                      'safflower',
                      'floudner',
                      'clay',
                      'duck',
                      'boar',
                      'turmeric',
                      'black pepper',
                      'ginger',
                      'goji',
                      'cherries',
                      'pomegranate',
                      'sage',
                      'basil',
                      'thyme',
                      'ashwagandha',
                      'reishi',
                      'turkey tail',
                      'shiitake',
                      'maitake',
                      'cordyceps',
                      'Fermentation Product',
                      'folate',
                      'mutton',
                      'mackerel',
                      'rabbit',
                      'bacon',
                      'cinnamon',
                      'fennel',
                      'avocado',
                      'apricot',
                      'artichoke',
                      'mushroom',
                      'almond',
                      'plasma',
                      'anchovy',
                      'sardine',
                      'fly larvae',
                      'butternut squash',
                      'zucchini',
                      'BHA',
                      'olive oil',
                      'fenugreek',
                      'quail',
                      'bok choy',
                      'hake',
                      'hoki oil'                      
                     ]


###############################################################################################################
#                                         Defined Functions:                                                  #
###############################################################################################################
###############################################################################################################
# Long list of allergens - takes all of the foods input by user and returns common ingredients                #
###############################################################################################################

def LongListAllergens(selectedFoods):
    filtered_df = df[df["name"].isin(selectedFoods)]


    filtered_df.drop('name', inplace = True, axis = 1)  # deleting working columns
    filtered_df.drop('url', inplace = True, axis = 1)
    filtered_df.drop('IngredientsDirty', inplace = True, axis = 1)
    filtered_df.drop('WIP', inplace = True, axis = 1)
    filtered_df.drop('IngredientsClean', inplace = True, axis = 1)
    filtered_df.drop('IngredientCount', inplace = True, axis = 1)
    filtered_df.drop_duplicates(inplace = True)  # removes duplicates, which cause errors downstream

    filtered_df = filtered_df.astype(int)  # turning boolean true / false into 1/0

    counter = 0
    toDrop = []

    for ingredient in ingredientUmbrella:  # finding ingredients with FALSE [0] on all foods
        if filtered_df[ingredient].sum() == 0:
            toDrop.append(ingredient)
 


#Dropping those columns
    filtered_df.drop(toDrop, inplace = True, axis = 1)

# adding food names back in
    filtered_df['foodName'] = selectedFoods

    cols_list = []  # defining list
    cols_list = filtered_df.columns.tolist()  # saving column names as own list
    cols_list = cols_list[:-1] # removing last item in list

        
# adding total row
    filtered_df.loc['Total'] = filtered_df.sum(numeric_only=True)

# sorting by total row
    filtered_df = filtered_df.sort_values('Total', axis = 1, ascending = False)
# dropping total row
    filtered_df = filtered_df.drop(index='Total', axis = 0)
    
    return filtered_df

###############################################################################################################
# Function: Visualize Allergens - creates graph shown in report tab                                           #
###############################################################################################################

def allergenVisualize(filtered_df, foodSubSet):

    plt.figure()
    plt.plot(filtered_df['foodName'], filtered_df[foodSubSet], marker='o', linestyle='-')
    plt.xticks(rotation=90,  wrap=True, fontsize=6)
    plt.yticks(ticks = [0,1], labels = ["Not Present", "Present"], rotation = 45, fontsize = 8)
    plt.legend(foodSubSet, loc = 'upper right')
    plt.title("Visualization of Potential Allergens")

    #plt.show()
    return plt.gcf()  # turns it into a figure that we can place in our canvas


###############################################################################################################
# Function: draw figure - allows you to actually print figure out in application                                                                          #
###############################################################################################################

def drawFigure(canvas, figure):
    if canvas.children:
        for child in canvas.winfo_children():
            child.destroy()  # destroys the child, hehe.  No actually it deletes the graph that's already there.
    
    figure_canvas_agg = FigureCanvasTkAgg(figure, canvas)
    
    figure_canvas_agg.draw()
    figure_canvas_agg.get_tk_widget().pack(side = 'top', fill = 'both', expand = 1)  # can change these for format
    return figure_canvas_agg
    
###############################################################################################################
# Function: Add food to database - allows users to input a food that is not currently in the db               #
###############################################################################################################  

def newFood(foodName, ingredientArray):
    df = pd.read_csv('ingredientDB.csv')

    # turning from string to list structure
    ingredientList = ingredientArray.split(', ')

    # counting the number of ingredients in each food
    IngredientCount= len(ingredientList)

    # creating array and then converting that array into a df
    newFoodArray = [foodName, None, None, ingredientArray, ingredientList, IngredientCount]
    newFoodDF = pd.DataFrame([newFoodArray], columns = ['name', 'url', 'IngredientsDirty', 'WIP', 'IngredientsClean', 'IngredientCount'])
                             
    # creating ingredient booleans
    for ingredient in ingredientUmbrella:   
        newFoodDF[ingredient] = newFoodDF['WIP'].str.contains(ingredient)

    df = pd.concat([df, newFoodDF])
    df.to_csv('IngredientDB.csv',index=False)

###############################################################################################################
# Function: calculate postieror probabilities for allergens                                                   #
############################################################################################################### 

def priorProbabilityAllergenCalc(longListAllergens):

    PriorProbsDict = {
        'beef': .36,
        'dairy': .28,
        'wheat': .15,
        'lamb':.06,
        'mutton':.06,
        'egg': .1,
        'chicken': .09,
        'soy': .06,
        'pork': .04,
        'rabbit':.01,
        }

  
    foods = ['beef', 'dairy', 'wheat', 'lamb', 'mutton', 'egg', 'chicken','soy','pork', 'rabbit']
    priorProbs = [.36, .28, .15, .06, .06, .1, .09, .06, .04, .01]
    
    df = longListAllergens[longListAllergens.columns.intersection(foods)]
    
    rows = df.shape[0]         
    cols = df.shape[1]
    
    colHeaders = df.columns.tolist()
    colHeaders = np.array(colHeaders)
    print(colHeaders)

    foodCount = df.sum(axis = 0).values # sum count
    foodCount = np.array(foodCount)

    foodTotal = [5]* cols
    foodTotal = np.array(foodTotal)

    probAllergy = [.36]* cols
    probAllergy = np.array(probAllergy)
    
    priorProbs = []

    for x in colHeaders:
        priorProbs.append(PriorProbsDict[x])
    
    priorProbs = np.array(priorProbs)

    
    bayes = (probAllergy * priorProbs) / (foodCount/foodTotal)
    bayes = np.round(bayes, 2)

    text = ''
    
    if len(bayes) ==0:
        text = "Research on allergens has not been done on the ingredients in your dog's problem foods.\nPlease use visualization tool to help identify common foods"
        return text
    else:
        for x in range(cols):
            text =  text + '\n' + 'The probability of your dog being allergic to ' + str(colHeaders[x]) + ' is ' + str(bayes[x])
            #print('The probability of your dog being allergic to ' + str(colHeaders[x]) + ' is' + str(bayes[x]))

        return text
    
    

###############################################################################################################
#                                     Tab 2 : Food Selector
###############################################################################################################
list2_0 = psg.Text('Search for Foods:')
list2_1 = psg.Text('Select where dog exhibits allergy symptoms, e.g. paw licking, hot spots, itchiness')

input2_0 = psg.Input('',
                     size = (25),
                     key = 'input2_0',
                     enable_events = True)  # Type food you are looking for


listbox2_0 = psg.Listbox(name, 
                         key = 'listbox2_0',
                         size = (70,8),
                         enable_events=True)  # filtered output to select foods

list2_3 = psg.Text('Foods to run report on')
button2_0 = psg.Button('Clear', key = 'button2_0')
button2_1 = psg.Button('Find Common Ingredients', key = 'button2_1')
listbox2_1 = psg.Multiline(key = 'listbox2_1',  # selected foods are displayed and stored here
                           size = (70,4),
                           enable_events=True)


tab2 = [[list2_1], [list2_0,input2_0], [listbox2_0],[list2_3, button2_0, button2_1], [listbox2_1]]


###############################################################################################################
#                                     Tab 3 : Welcome
###############################################################################################################
l1 = psg.Text('Welcome to the dog food allergen tool!')
l2 = psg.Text('This is meant to help dog owners identify possible food allergens for their pets,\n and suppy them with potential next foods to try.')  
l3 = psg.Text("\nPlease consult your veterinarian before making changes to your dog's diet")

tab3 = [[l1],[l2],[l3]]


###############################################################################################################
#                                     Tab 4 : Report Visualization
###############################################################################################################
list4_0 = psg.Text('Visual Report:')
list4_1 = psg.Text("These are the common ingredinets in your dog's problem foods")

listbox4_0 = psg.Listbox('empty', key = 'listbox4_0',  # selected foods are displayed and stored here
                           size = (15,10),
                           enable_events=True,
                           select_mode = psg.LISTBOX_SELECT_MODE_MULTIPLE)

list4_2 = psg.Text('Find kibbles that do not include selected ingredients')
button4_0 = psg.Button('Find Kibbles', key = 'button4_0',enable_events = True)

canvas4_0 = psg.Canvas(key='canvas4_0')
                       #, size = (5, 5))


tab4 = [[list4_0], [list4_1], [listbox4_0, canvas4_0], [list4_2, button4_0]]

###############################################################################################################
#                                     Tab 5 : Add Food
###############################################################################################################
right_click_menu = ['', ['Paste']]

list5_0 = psg.Text('If you can not find a food you are looking for, you may add it manually here.')
list5_1 = psg.Text('- Please pull up the food you would like to add on Chewy.com.\n- For the food name, copy the whole name as it appears but leave off any information about\n  food bag size, e.g. "35 lb bag".\n-For the ingredients, please copy and paste in the entire list of ingredients as it \n  appears on Chewy.com')
list5_2 = psg.Text('Food Name')
input5_1 = psg.Input('',
                     size = (70),
                     key = 'input5_1',
                     enable_events = True,
                     do_not_clear = False,
                     right_click_menu = right_click_menu)  # Type food you are adding to db

list5_3 = psg.Text('Ingredients')
input5_2 = psg.Multiline('',
                     size = (70,4),
                     key = 'input5_2',
                     enable_events = True,
                     do_not_clear=False,
                     right_click_menu = right_click_menu)  # Type food you are looking for
button5_0 = psg.Button('Add to Database', 
                       key = 'button5_0')

button5_1 = psg.Button('Paste Food Name', 
                       key = 'button5_1')
button5_2 = psg.Button('Paste Ingredients', 
                       key = 'button5_2')

tab5 = [[list5_0], [list5_1], [list5_2, button5_1], [input5_1], [list5_3, button5_2], [input5_2], [button5_0]]

###############################################################################################################
#                                     Tab 6 : posterior probs and food elimination
###############################################################################################################
list6_0 = psg.Text('Report:')
list6_1 = psg.Text("The following foods do not contain: ", key = 'list6_1')

multiline6_0 = psg.Listbox('', size = (70,7), key = 'multiline6_0')

list6_2 = psg.Text("Using Bayes theorum the probability of your dog being allergic to the following ingredients \nis as follows.")
list6_3 = psg.Text('Note: This is only calculated for beef, wheat, egg, chicken, soy, pork, rabbit, lamb, and mutton')

multiline6_1 = psg.Multiline('', size = (70,3), key = 'multiline6_1')

tab6 = [[list6_0], [list6_1], [multiline6_0], [list6_2], [list6_3],[multiline6_1]]

###############################################################################################################
#                                              Layout / setup
###############################################################################################################

layout = [[psg.TabGroup([[

    psg.Tab('Welcome', tab3),
    psg.Tab('Food Selector', tab2),
    psg.Tab('Common Ingredients', tab4),
    psg.Tab('Report', tab6),
    psg.Tab('Add Food to Database', tab5)]])]]


window = psg.Window('Godbole Capstone', layout, finalize = True) # finalize = true for graph, allows you to access tkinter functionality


###############################################################################################################
#                                              Events
###############################################################################################################

# Event Loop to 
while True:
    event, values = window.read()
    if event in (psg.WIN_CLOSED, 'Exit'):                # always check for closed window
        break
        
    if values['input2_0'] != '':                         # if a keystroke entered in food search
        search = values['input2_0']
        new_values = [x for x in name if search in x]  # do the filtering
        window['listbox2_0'].update(new_values)     # display in the listbox
    else:
        # display original unfiltered list
        window['listbox2_0'].update(name)
        
        
    # if a list item is chosen
    if event == 'listbox2_0' and len(values['listbox2_0']):
        allergicFoods.append(values['listbox2_0'])
        window['listbox2_1'].update(allergicFoods)
        
    if event == 'button2_0':
        allergicFoods = []
        window['listbox2_1'].update(allergicFoods)
    if event =='button2_1':
        allergicFoods = list(map(str, chain.from_iterable(allergicFoods)))  # turns food list from a list of lists to just a list
        
    
        filtered_df = LongListAllergens(allergicFoods)  # creates a new dataframe per report of foods / ingredients
        ingredientList =[]  # defining list, deletes old lists
        ingredientList = filtered_df.columns.tolist()  # saving column names as own list
        ingredientList = ingredientList[:-1] # removing last item in list
        window['listbox4_0'].update(ingredientList)
    
    if event == 'listbox4_0':
        selection = values[event]
        
        #canvas.delete('all')
        drawFigure(window['canvas4_0'].TKCanvas, allergenVisualize(filtered_df, selection))

    if event =='button5_0':
        #name = values['input5_1']
        #ingredients = value['input5_2']
        newFood(name, ingredients) # running newFood definition
        df = pd.read_csv('ingredientDB.csv')
        name = df['name']
        window['listbox2_0'].update(name)
        
        
    if event=='button5_1':
         text = psg.clipboard_get()
         textinput2 = values['input5_2']
         window['input5_1'].update(text)
         window['input5_2'].update(textinput2)
         name = text
    if event=='button5_2': 
         text = psg.clipboard_get()
         textinput1 = values['input5_1']
         window['input5_1'].update(textinput1)
         window['input5_2'].update(text)
         ingredients = text

    if event == 'button4_0':
        window['multiline6_0'].update([])
        ingredientExclude = window['listbox4_0'].get()  # getting elements from ingredient selection
        ingredientExcludeList = ', '.join(ingredientExclude) # turning list to string
        excludeDF = df
        
        for x in ingredientExclude:
            excludeDF.drop(excludeDF[excludeDF[x]==True].index, inplace = True)
            
        print(excludeDF.shape)
        
        GoodFoods = excludeDF['name'].tolist()
        displayText = "These foods do not contain the following ingredients: " + ingredientExcludeList
        window['list6_1'].update(displayText)

        window['multiline6_0'].update(GoodFoods)
        print(ingredientExclude)
        #filtered_df = LongListAllergens(allergicFoods)  # creates a new dataframe per report of foods / ingredients
 
        bayesText =  priorProbabilityAllergenCalc(filtered_df)

        window['multiline6_1'].update(bayesText)

    
window.close()
```
